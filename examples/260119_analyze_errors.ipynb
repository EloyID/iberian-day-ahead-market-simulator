```python
import sys
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
FILE_DIR = os.getcwd()
# parent directory and src directory
SRC_DIR = os.path.join(os.path.dirname(FILE_DIR), "src")
DATA_DIR = os.path.join(os.path.dirname(FILE_DIR), "data")
print(f"Adding src directory to sys.path: {SRC_DIR}")
# Add the src directory to sys.path
sys.path.append(SRC_DIR)
```

```python
from mibel_simulator.const import RDC_ENERGY_COLUMNS, RDC_PRICE_COLUMNS
from mibel_simulator.residual_demand_curve import interpolate_residual_demand_curves


START_DATE = "2025-06-15"  # "2025-06-15"
END_DATE = "2025-07-15"

RDC_ANALYTICS_PATTERN = "results/rdc_analytics_{YYYYMMDD}.parquet"
RDC_PV_PATTERN = "results/residual_demand_curve_pv_{YYYYMMDD}.parquet"
RDC_HYDRO_PATTERN = "results/residual_demand_curve_hydro_{YYYYMMDD}.parquet"
RDC_BASIC_PATTERN = "results/residual_demand_curve_basic_{YYYYMMDD}.parquet"


ONLY_SIMPLE_RDC_ENERGY_COLUMNS = [
    f"only_simple_rdc_{col}" for col in RDC_ENERGY_COLUMNS
]
ONLY_SIMPLE_RDC_PRICE_COLUMNS = [f"only_simple_rdc_{col}" for col in RDC_PRICE_COLUMNS]
ONLY_SIMPLE_RDC_PRICE_ERROR_COLUMNS = [
    f"only_simple_rdc_price_error_{col}" for col in RDC_PRICE_COLUMNS
]
SUBMITTED_RELAXED_RDC_ENERGY_COLUMNS = [
    f"submitted_relaxed_rdc_{col}" for col in RDC_ENERGY_COLUMNS
]
SUBMITTED_RELAXED_RDC_PRICE_COLUMNS = [
    f"submitted_relaxed_rdc_{col}" for col in RDC_PRICE_COLUMNS
]
SUBMITTED_RELAXED_RDC_PRICE_ERROR_COLUMNS = [
    f"submitted_relaxed_rdc_price_error_{col}" for col in RDC_PRICE_COLUMNS
]
COMPLEX_I_RDC_ENERGY_COLUMNS = [f"complex_I_rdc_{col}" for col in RDC_ENERGY_COLUMNS]
COMPLEX_I_RDC_PRICE_COLUMNS = [f"complex_I_rdc_{col}" for col in RDC_PRICE_COLUMNS]
COMPLEX_I_RDC_PRICE_ERROR_COLUMNS = [
    f"complex_I_rdc_price_error_{col}" for col in RDC_PRICE_COLUMNS
]
COMPLEX_II_RDC_ENERGY_COLUMNS = [f"complex_II_rdc_{col}" for col in RDC_ENERGY_COLUMNS]
COMPLEX_II_RDC_PRICE_COLUMNS = [f"complex_II_rdc_{col}" for col in RDC_PRICE_COLUMNS]
COMPLEX_II_RDC_PRICE_ERROR_COLUMNS = [
    f"complex_II_rdc_price_error_{col}" for col in RDC_PRICE_COLUMNS
]

analytic_rdc_columns = {
    "only_simple_rdc": {
        "energy": ONLY_SIMPLE_RDC_ENERGY_COLUMNS,
        "price": ONLY_SIMPLE_RDC_PRICE_COLUMNS,
        "price_error": ONLY_SIMPLE_RDC_PRICE_ERROR_COLUMNS,
    },
    "submitted_relaxed_rdc": {
        "energy": SUBMITTED_RELAXED_RDC_ENERGY_COLUMNS,
        "price": SUBMITTED_RELAXED_RDC_PRICE_COLUMNS,
        "price_error": SUBMITTED_RELAXED_RDC_PRICE_ERROR_COLUMNS,
    },
    "complex_I_rdc": {
        "energy": COMPLEX_I_RDC_ENERGY_COLUMNS,
        "price": COMPLEX_I_RDC_PRICE_COLUMNS,
        "price_error": COMPLEX_I_RDC_PRICE_ERROR_COLUMNS,
    },
    "complex_II_rdc": {
        "energy": COMPLEX_II_RDC_ENERGY_COLUMNS,
        "price": COMPLEX_II_RDC_PRICE_COLUMNS,
        "price_error": COMPLEX_II_RDC_PRICE_ERROR_COLUMNS,
    },
}

analytics_rdcs = [
    "only_simple_rdc",
    "submitted_relaxed_rdc",
    "complex_I_rdc",
    "complex_II_rdc",
]

dfs_dict = {"rdc_pv": {}, "rdc_hydro": {}, "rdc_basic": {}}
```

```python
for date in pd.date_range(START_DATE, END_DATE):
    try:
        rdc_analytics = pd.read_parquet(
            RDC_ANALYTICS_PATTERN.replace("{YYYYMMDD}", date.strftime("%Y%m%d"))
        )
        rdc_pv = pd.read_parquet(
            RDC_PV_PATTERN.replace("{YYYYMMDD}", date.strftime("%Y-%m-%d"))
        )
        rdc_hydro = pd.read_parquet(
            RDC_HYDRO_PATTERN.replace("{YYYYMMDD}", date.strftime("%Y-%m-%d"))
        )
        rdc_basic = pd.read_parquet(
            RDC_BASIC_PATTERN.replace("{YYYYMMDD}", date.strftime("%Y-%m-%d"))
        )

        optimized_rdcs = [rdc_pv, rdc_hydro, rdc_basic]
        optimized_rdcs_list = ["rdc_pv", "rdc_hydro", "rdc_basic"]

        for rdc_df, rdc_name in zip(optimized_rdcs, optimized_rdcs_list):
            for analytics_columns_dict in analytic_rdc_columns.values():
                rdc_df[analytics_columns_dict["price"]] = (
                    interpolate_residual_demand_curves(
                        target_energy_levels=rdc_df[RDC_ENERGY_COLUMNS],
                        residual_demand_curves=rdc_analytics[
                            RDC_PRICE_COLUMNS + analytics_columns_dict["energy"]
                        ],
                        residual_demand_energy_columns=analytics_columns_dict["energy"],
                        extrapolate_action="nan",
                    )[RDC_PRICE_COLUMNS]
                )

                rdc_df[analytics_columns_dict["price_error"]] = (
                    rdc_df[analytics_columns_dict["price"]].values
                    - rdc_df[RDC_PRICE_COLUMNS].values
                )

            dfs_dict[rdc_name][date.strftime("%Y-%m-%d")] = rdc_df

    except Exception as e:
        print(f"Error processing date {date.strftime('%Y-%m-%d')}: {e}")
        continue
```
